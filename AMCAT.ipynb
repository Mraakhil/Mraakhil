````python
fig, axs = plt.subplots(1, 2, figsize=(8, 5), layout="constrained")
fig.suptitle("Salary distribution with Outliers")

axs[0].hist(numeric_columns['Salary'], density=True)
numeric_columns['Salary'].plot(kind = 'kde', ax = axs[0])
axs[0].set_title("Salary", fontsize="medium")

axs[1].boxplot(numeric_columns["Salary"])
axs[1].set_title("Salary", fontsize="medium")

plt.show()

````

````python
Q1 = numeric_columns['Salary'].quantile(0.25)
Q3 = numeric_columns['Salary'].quantile(0.75)

IQR = Q3 - Q1

l1 = (Q1 * 1.5 - IQR)
l2 = (Q3 * 1.5 + IQR)
````

````python
salary_no_outliers = numeric_columns[(numeric_columns['Salary'] > l1) & (numeric_columns['Salary'] < l2)]

salary_no_outliers
````
