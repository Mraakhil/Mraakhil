```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

final = pd.DataFrame()  # Initialize an empty DataFrame

for j in range(1,15):  # Loop through pages 1 to 9
    url = 'https://www.oyorooms.com/hotels-in-hyderabad/?page=' + format(j)
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')

    # Initialize lists for each page
    name = []
    price = []
    allinfo = []
    rating = []
    facilities = []

    # Scrape data for the current page
    for i in soup.find_all('h3'):
        name.append(i.text.strip())
    for i in soup.find_all('span', class_='listingPrice__finalPrice listingPrice__finalPrice--black'):
        price.append(i.text.strip())
    for i in soup.find_all('div', class_='listingPrice__numbers'):
        allinfo.append(i.text.strip())
    for i in soup.find_all('div', class_='hotelRating'):
        rating.append(i.text.strip())
    for i in soup.find_all('div', class_='amenityWrapper'):
        facilities.append(i.text.strip())

    # Create a DataFrame for the current page
    df = pd.DataFrame({
        'name': name,
        'price': price,
        'allinfo': allinfo,
        'rating': rating,
        'facilities': facilities
    })

    # Concatenate the current page data to the final DataFrame
    final = pd.concat([final, df], ignore_index=True)

# Now `final` contains the scraped data from all pages.
```
